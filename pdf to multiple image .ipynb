# pdf to image for one file

```python
#The pdf2image library can be used
#You can install it simply using,

#pip install pdf2image

#Once installed you can use following code to get images.

from pdf2image import convert_from_path
pages = convert_from_path(r"C:\Users\siva\Downloads\B655_V1983_2060282.PDF", 300)

#Saving pages in jpeg forma\
for i, image in enumerate(pages):
    fname = 'image'+str(i)+'.png'
    image.save(fname, "PNG")    
```

# pdf to images for more than one file

```python
from pdf2image import convert_from_path
import glob,os
import os, subprocess

pdf_dir = r'C:\Users\siva\Desktop\arshiya\remedy'
os.chdir(pdf_dir)

for pdf_file in glob.glob(os.path.join(pdf_dir, "*.pdf")):
    pages = convert_from_path(pdf_file, 500)
    for i, page in enumerate(pages):
        page.save(pdf_file[:-4] +str(i)+".jpg", 'JPEG')
```
